```python
# ===============================
#  1. SETUP + VERIFY PATHS
# ===============================
import os, json, cv2, glob
from detectron2.data.datasets import register_coco_instances
from detectron2.engine import DefaultTrainer
from detectron2.config import get_cfg
from detectron2 import model_zoo
from detectron2.data import DatasetMapper, build_detection_train_loader
from detectron2.data import transforms as T

dataset_root = "/content/dataset"
train_json = os.path.join(dataset_root, "train", "_annotations.coco.json")
val_json   = os.path.join(dataset_root, "valid", "_annotations.coco.json")

# ===============================
#  2. FILTER TO ONLY POTHOLE CLASS
# ===============================
def filter_pothole_only(json_path, output_path):
    with open(json_path) as f:
        data = json.load(f)

    pothole_category_id = None
    for cat in data["categories"]:
        if cat["name"].lower() == "pothole":
            pothole_category_id = cat["id"]

    if pothole_category_id is None:
        raise ValueError("No 'pothole' category found in dataset!")

    # Keep only pothole annotations
    data["annotations"] = [a for a in data["annotations"] if a["category_id"] == pothole_category_id]
    data["categories"] = [c for c in data["categories"] if c["id"] == pothole_category_id]

    with open(output_path, "w") as f:
        json.dump(data, f)
    print(f" Saved filtered pothole dataset: {output_path}")

filter_pothole_only(train_json, os.path.join(dataset_root, "train", "_annotations_pothole.coco.json"))
filter_pothole_only(val_json, os.path.join(dataset_root, "valid", "_annotations_pothole.coco.json"))

# ===============================
#  3. APPLY CLAHE ENHANCEMENT
# ===============================
def apply_CLAHE_to_folder(folder_path):
    image_paths = glob.glob(os.path.join(folder_path, "*.jpg")) + glob.glob(os.path.join(folder_path, "*.png"))
    print(f"Applying CLAHE to {len(image_paths)} images in {folder_path}")
    for img_path in image_paths:
        img = cv2.imread(img_path)
        if img is None:
            continue
        lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
        l, a, b = cv2.split(lab)
        clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
        cl = clahe.apply(l)
        merged = cv2.merge((cl, a, b))
        enhanced = cv2.cvtColor(merged, cv2.COLOR_LAB2BGR)
        cv2.imwrite(img_path, enhanced)

apply_CLAHE_to_folder(os.path.join(dataset_root, "train"))
apply_CLAHE_to_folder(os.path.join(dataset_root, "valid"))

# ===============================
#  4. CUSTOM TRAINER WITH AUGMENTATIONS
# ===============================
class MyAugmentedTrainer(DefaultTrainer):
    @classmethod
    def build_train_loader(cls, cfg):
        augmentations = [
            T.RandomBrightness(0.7, 1.3),
            T.RandomContrast(0.8, 1.3),
            T.RandomSaturation(0.8, 1.3),
            T.RandomRotation(angle=[-20, 20]),
            T.RandomFlip(prob=0.5, horizontal=True, vertical=False),
            T.RandomLighting(0.7),
            T.RandomCrop("relative_range", (0.9, 0.9)),
        ]
        return build_detection_train_loader(
            cfg, mapper=DatasetMapper(cfg, is_train=True, augmentations=augmentations)
        )

# ===============================
#  5. DETECTRON2 CONFIGURATION
# ===============================
register_coco_instances("road_pothole_train", {},
                        os.path.join(dataset_root, "train", "_annotations_pothole.coco.json"),
                        os.path.join(dataset_root, "train"))
register_coco_instances("road_pothole_val", {},
                        os.path.join(dataset_root, "valid", "_annotations_pothole.coco.json"),
                        os.path.join(dataset_root, "valid"))

cfg = get_cfg()
cfg.merge_from_file(model_zoo.get_config_file("COCO-Detection/faster_rcnn_R_50_FPN_3x.yaml"))
cfg.DATASETS.TRAIN = ("road_pothole_train",)
cfg.DATASETS.TEST = ("road_pothole_val",)
cfg.DATALOADER.NUM_WORKERS = 2

cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url("COCO-Detection/faster_rcnn_R_50_FPN_3x.yaml")
cfg.MODEL.ROI_HEADS.NUM_CLASSES = 1  # pothole only
cfg.SOLVER.IMS_PER_BATCH = 2
cfg.SOLVER.BASE_LR = 0.00025
cfg.SOLVER.MAX_ITER = 15000
cfg.SOLVER.STEPS = (10000, 13000)
cfg.SOLVER.CHECKPOINT_PERIOD = 1000
cfg.OUTPUT_DIR = "./output_pothole"
os.makedirs(cfg.OUTPUT_DIR, exist_ok=True)

# ===============================
#  6. TRAIN MODEL
# ===============================
trainer = MyAugmentedTrainer(cfg)
trainer.resume_or_load(resume=False)
trainer.train()
```

```python
from detectron2.evaluation import COCOEvaluator, inference_on_dataset
from detectron2.data import build_detection_test_loader

evaluator = COCOEvaluator("road_pothole_val", cfg, False, output_dir="./output_pothole/")
val_loader = build_detection_test_loader(cfg, "road_pothole_val")
print(inference_on_dataset(trainer.model, val_loader, evaluator))
```

```python
# Reimport all necessary modules
import cv2, torch, random, glob
from detectron2.engine import DefaultPredictor
from detectron2.config import get_cfg
from detectron2.utils.visualizer import Visualizer
from detectron2.data import MetadataCatalog
from detectron2 import model_zoo
import matplotlib.pyplot as plt
import os

# Recreate your config and load weights
cfg = get_cfg()
cfg.merge_from_file(model_zoo.get_config_file("COCO-Detection/faster_rcnn_R_50_FPN_3x.yaml"))
cfg.MODEL.WEIGHTS = "./output_pothole/model_final.pth"  # path to your trained model
cfg.MODEL.ROI_HEADS.NUM_CLASSES = 1
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.5
cfg.DATASETS.TEST = ("road_pothole_val",)

# Create predictor again
predictor = DefaultPredictor(cfg)
```

```python
val_images = glob.glob("/content/dataset/valid/*.jpg") + glob.glob("/content/dataset/valid/*.png")
random_image = random.choice(val_images)
print(f"Testing on random image: {random_image}")

im = cv2.imread(random_image)
outputs = predictor(im)

v = Visualizer(im[:, :, ::-1], MetadataCatalog.get("road_pothole_val"), scale=0.8)
out = v.draw_instance_predictions(outputs["instances"].to("cpu"))

plt.figure(figsize=(12, 8))
plt.imshow(out.get_image()[:, :, ::-1])
plt.axis("off")
plt.show()
```

```python
from transformers import DetrForObjectDetection, DetrImageProcessor
import torch, requests
from PIL import Image

processor = DetrImageProcessor.from_pretrained("facebook/detr-resnet-50")
model = DetrForObjectDetection.from_pretrained("facebook/detr-resnet-50")

# Example with one of your road images
image_path = "/content/dataset/train/verypoor_126_jpg.rf.a2b34e6a6fd775b4719ffa5a786e1f5a.jpg"
image = Image.open(image_path).convert("RGB")

inputs = processor(images=image, return_tensors="pt")
outputs = model(**inputs)

# Postprocess results
target_sizes = torch.tensor([image.size[::-1]])
results = processor.post_process_object_detection(outputs, target_sizes=target_sizes)[0]

for score, label, box in zip(results["scores"], results["labels"], results["boxes"]):
    if score > 0.5:
        print(f"{model.config.id2label[label.item()]}: {round(score.item(), 3)} at {box.tolist()}")
```
